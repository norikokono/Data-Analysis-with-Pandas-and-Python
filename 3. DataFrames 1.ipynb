# The .astype() Method

```python
# astpe() the method requires a Series to not have any null vaalues.
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"].fillna(0, inplace = True)
nba["College"].fillna("None", inplace = True)
nba.head(3)
```

```python
nba.dtypes
```

```python
# Object is just pandas internal syntax for string.
# Float is important caveat, which is that in our original CSV, all of these numeric columns are actually stored as 
# integers, and when pandas imports them and sees these null cells, it hahas to convert the columns to floats.
# That's just something technical it has to do on the backend to store it.
```

```python
nba.info()
```

```python
# converting from more complex data type like floats to intergers, you can actually see a memory decerease because 
# it's easier to store an interger.
```

```python
nba["Salary"] = nba["Salary"].astype("int") # integer
```

```python
nba["Number"] = nba["Number"].astype("int")
```

```python
# category Only for Pandas, but it doesn't exist in Python
# A small number of unique values in a DataFrame. e.q. genger, Reduce momory usage
nba["Position"].nunique()
```

```python
nba["Popsition"] = nba["Position"].astype("category")
```

```python
nba["Team"].nunique()
```

```python
nba["Team"] = nba["Team"].astype("category")
```

```python
nba.head()
```

```python
nba.info()
# memory usage was decliend to 34.5+ KB
```

# Rank Values with the .rank() Method

```python
nba = pd.read_csv("nba.csv").dropna(how = "all")
nba["Salary"] = nba["Salary"].fillna(0).astype("int") # It still have to call the astype method to convert them to integers.
nba.head(3)
```

```python
# rank Method 
nba["Salary"].rank()
```

```python
nba["Salary Rank"] = nba["Salary"].rank(ascending = False).astype("int")
nba.head(3)
```

```python
nba.sort_values(by = "Salary", ascending = False)
```
