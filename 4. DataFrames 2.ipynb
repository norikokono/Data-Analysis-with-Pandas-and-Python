# The .isin() Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
# .isin() Method can be useful for checking for muliple values withon a single series.
# This is especially helpful when you want to complete a filter process that involves multiple values and
# you don't want to create multiple boolean series.
```

```python
mask1 = df['Team'] == "Legal"
mask2 = df["Team"] == "Sales"
mask3 = df["Team"] == "Product"

df[mask1 | mask2 | mask3]
# This is inefficient, so it is better to use .isin() Method.
```

```python
df["Team"].isin(["Legal", "Sales", "Product"])
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product"])
df[mask]
```

```python
mask = df["Team"].isin(["Legal", "Sales", "Product", 'Marketing'])
df[mask]
```

# The .duplicated() Method

```python
df = pd.read_csv("employees.csv", parse_dates = ['Start Date', "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True) # permanenly modify the DataFrame
df.head(3)
```

```python
df["First Name"]
```

```python
# .duplicated(subset=None, keep='first')   
# keep: By default, keep marks the very first occurrence of each values as a non duplicate.
df["First Name"].duplicated()
```

```python
df[df['First Name'].duplicated()]
```

```python
df["First Name"].duplicated(keep = "last")
```

```python
df[df["First Name"].duplicated(keep = "last")]
```

```python
df["First Name"].duplicated(keep = False) # Mark something as duplicated if it occurs more than once
```

```python
df[df["First Name"].duplicated(keep = False)]
```

```python
# Remove all of rows that are unique
# A boolean series to negate it or convert it the other way. Use Tilde ~
# All of the True will become False and all of the False will become True.
~df["First Name"].duplicated(keep = False)
```

```python
mask = ~df["First Name"].duplicated(keep = False)  
df[mask] # This boolean Series represents the unique values.
```

# The .drop_duplicates() Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
# Python bulit in len() Function
len(df) # the number of rows
```

```python
df.drop_duplicates()
```

```python
len(df.drop_duplicates())
# There is still 1,000 rows even after the drop_duplicates() Method has been caled. Why is this?
# The reason is because it doesn't matter if a duplicate value occurs in a single column.
# There is no two rows where the values are exactly identical.
```

```python
# .drop_duplicates(subset=None) subset is set to none by default.
# subset accepts a list of Strings, and those Strings can represent the names of columns in which we want to check for duplicates.
df.drop_duplicates(subset = ["First Name"], keep = "first") # default: keep = 'first'
```

```python
df.drop_duplicates(subset = ["First Name"], keep = "last")  # The last occurrence of each duplicate
```

```python
df.drop_duplicates(subset = ["First Name"], keep = False)
```

```python
df.drop_duplicates(subset = ["Team"], keep = False)
# We got an empty DataFrame because there is no single Team value that only occurs once in that Team column.
# A keep argument is going to completely wipe them out.
```

```python
df.drop_duplicates(subset = ["First Name", "Team"], inplace = True)
```
