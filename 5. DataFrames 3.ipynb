# The .apply() Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
# index 0: 1985, index 1: Roger Moore, index 2: John Glen, index 3: 275.2
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns") # traversing the columns as we apply this operation for every row
```
